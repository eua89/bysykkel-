```python
from BeautifulSoup import BeautifulSoup
import urllib

url = 'https://developer.oslobysykkel.no/data'
r = urllib.urlopen(url).read()
soup = BeautifulSoup(r)
print(soup.prettify())
```

```python
values= [row.get('href') for row in soup.findAll('a', {"class":'button'})]
print values
```

```python
import re
def make_url(ending):
    beginning ='https://developer.oslobysykkel.no/data'
    return beginning+'/'+ending.split('/')[2]

total_link=[make_url(ending) for ending in values if re.search(".json", ending)]
print total_link
```

```python
import wget

for i,link in enumerate(total_link):
    wget.download(link, 'file_{}.zip'.format(i))
    
    
```

```python
import os,zipfile

dir_name = os.getcwd()+'/'
extension = '.zip'

for item in os.listdir(dir_name):
    if item.endswith(extension):
        file_name = dir_name + item
        try:
            zip_ref = zipfile.ZipFile(file_name) 
            zip_ref.extractall(dir_name) 
            zip_ref.close() 
            os.remove(file_name) 
        except:
            pass        
```
